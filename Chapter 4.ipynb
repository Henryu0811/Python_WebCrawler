#### 用 requests 模擬瀏覽器運作的方式，透過 get 和 post 的方法傳送 request，將回傳的資訊交由 BeautifulSoup 解析後，取得 HTML裡面元素的資料。

```python
import requests
from bs4 import BeautifulSoup

# 用 get() 函式連結網頁，取得網頁資料
# http 通訊協定的溝通方式，瀏覽器傳送 request，主機回傳 response
resp = requests.get('https://code-gym.github.io/spider_demo/') # 這一行回傳的資料就是 response
# 將 response 交給 BeautifulSoup 解析
soup = BeautifulSoup(resp.text, 'html5lib') # .text 代表取得 response 中的文字內容，html5lib 是分析器的名稱
# 用函式 find() 尋找
print(soup.find('h1').text)
print(soup.h1.text)
```

## find_all

```python
for h3 in soup.find_all('h3'):
    print(h3.a) # .a，h3 後面有超連結的會顯示，沒有的顯示 None
```

## 使用 class 取得網頁中元素資料

```python
# find_all(標籤名稱, 標籤屬性)
for title in soup.find_all('h3', 'post-title'):
    print(title.a)
```

## 使用 key-value 取得網頁中元素資料

找文章中的分類標籤，class 屬性有兩個值，class="post-category cat-1" href="category.html">Web Design

```python
# find_all(標籤名稱, 標籤屬性)
for cat in soup.find_all('a', {'class': 'post-category', 'class': 'cat-1'}):
    print(cat)
```

## 抓取部落格文章標題和資訊
    ● stripped_strings

```python
for posts in soup.find_all('div', 'post-body'):
    for post in posts.stripped_strings:
        print(post)
```

## PTT NBA
https://www.ptt.cc/bbs/NBA/index.html

    ● 所有"今天"發的"文章標題"
    ● "推文數"超過10

```python
import requests
from bs4 import BeautifulSoup
import time

today = time.strftime('%m/%d').lstrip('0') # 回傳的月份前面會有 0 ，所以用 lstrip() 函式，把左邊的文字移除

def pttNBA(url):
    resp = requests.get(url) # 連上 ptt 的 NBA 版，取得回傳後的 response
    # 實戰中，server 是會停止運作的，或是更改 url 路徑，所以要先確定 server 是否正確回應網頁內容
    if resp.status_code != 200: # 用 response 的 status_code 判斷
        print('URL發生錯誤： ' + url)
        return # 發生錯誤就列印錯誤訊息，並且 return 跳出函式

    soup = BeautifulSoup(resp.text, 'html5lib') # 將 response 的內容傳給 BeautifulSoup 解析
    # 取得上一頁的連結
    # 上一頁的連結順序在第二個，所以取索引值 1 的資料
    # ['href']，直接取得屬性中的資料
    paging = soup.find('div', 'btn-group btn-group-paging').find_all('a')[1]['href']
    
    articles = []
    # 文章都是用標籤名稱 div 和 class 名稱 r-ent包起來的
    rents = soup.find_all('div', 'r-ent')
    for rent in rents:
        title = rent.find('div', 'title').text.strip() # strip() 去掉空白字元
        count = rent.find('div', 'nrec').text.strip()
        date = rent.find('div', 'meta').find('div', 'date').text.strip()
        # 這樣好像也可以
        # date = rent.find('div', 'date').text.strip()
        article = '%s %s:%s' % (date, count, title) # 抓到的資料，整理成一個字串 article
        
        # 判斷文章是不是今天發的，發文數有沒有超過 10
        # 用 try-except 包 if 判斷式，因為將推文數字串轉為 int 的過程，如果遇到 '爆'或是'X'，會出錯
        try:
            if today == date and int(count) > 10:
                articles.append(article)
        except:
            if today == date and count == '爆':
                articles.append(article)
                
    if len(articles) != 0:
        for article in articles:
            print(article)
        pttNBA('https://www.ptt.cc' + paging) # 用上一頁的連結呼叫 pttNBA 函式，繼續找文章
    else:
        return

# 呼叫 pttNBA 函式開始找資料
pttNBA('https://www.ptt.cc/bbs/NBA/index.html')
```

    ● 抓取"前五頁""文章標題"
    ● "推文數"超過10

```python
import requests
from bs4 import BeautifulSoup
import time


def pttNBA(url):
    resp = requests.get(url)
    if resp.status_code != 200:
        print('URL發生錯誤：' + url)
        return
    
    soup = BeautifulSoup(resp.text, 'html5lib')
    paging = soup.find('div', 'btn-group btn-group-paging').find_all('a')[1]['href']
    
    articles = []
    rents = soup.find_all('div', 'r-ent')
    for rent in rents:
        title = rent.find('div', 'title').text.strip()
        count = rent.find('div', 'nrec').text.strip()
        date = rent.find('div', 'meta').find('div', 'date').text.strip()
        article = '%s %s:%s' % (date, count, title)
        
    for i in range(0, 3):
        try:
            if int(count) > 50:
                articles.append(article)
        except:
            if count == '爆':
                articles.append(article)

        for article in articles:
            print(article)
        if i < 2:
            pttNBA('https://www.ptt.cc' + paging)
        else:
            break
    
pttNBA('https://www.ptt.cc/bbs/NBA/index.html')
```

## 台鐵時刻表
https://tip.railway.gov.tw/tra-tip-web/tip

##### post
查詢車次：輸入資訊都被 form 標籤包起來，裡面有個屬性 method="post" 
http 通訊協定會向 server提出請求，請求的方式有 get 跟 post，一般向主機送出參數是用 psot，也能使用 get ，在網址後加上問號，傳送參數，
但用 get 的網址參數是明碼，會被有心人士利用參數資訊，所以一般身分驗證輸入的內容，用 post 方法向主機傳送 request

##### CSRF (Cross-site request forgery) 跨站請求偽造
是一個欺騙網路使用者的技術

```python
import requests
from bs4 import BeautifulSoup
import time

url = 'https://tip.railway.gov.tw/tra-tip-web/tip'
staDic = {} # 儲存找到的火車站名稱和代碼
today = time.strftime('%Y/%m/%d')
sTime = '06:00'
eTime = '12:00'

def getTrip():
    # 透過 requests 取得主機回傳的 response
    resp = requests.get(url)
    # 確認回傳代碼是否成功
    if resp.status_code !=200:
        print('URL發生錯誤： ' + url)
        return
    
    # 解析 HTML 原始碼，取得 BeautifulSoup 物件
    soup = BeautifulSoup(resp.text, 'html5lib')
    # 先找到屬性 id 為 cityHot 的標籤，裡面有 ul 標籤，直接用 .ul，如果只有一個 li 也可以直接 .li，但因為有多個，所以用 find_all
    stations = soup.find(id = 'cityHot').ul.find_all('li')
    for station in stations:
        stationName = station.button.text
        stationId = station.button['title'] # 用 []  直接取得值
        staDic[stationName] = stationId
        
    csrf = soup.find(id = 'queryForm').find('input', {'name': '_csrf'})['value'] # 取得 csrf 代碼
    # 要傳給主機查詢車次時刻表的表單
    formData = {
        'trainTypeList': 'ALL',
        'transfer': 'ONE',
        'startOrEndTime': 'true',
        'startStation': staDic['臺北'],
        'endStation': staDic['新竹'],
        'rideDate': today,
        'startTime': sTime,
        'endTime': eTime
    }
    
    queryUrl = soup.find(id='queryForm')['action']
    # 用 requests.post() 傳送表單給主機
    # https://tip.railway.gov.tw，這是台鐵 domain，也就是台鐵網域名稱
    # 跟取得首頁資料不同，這是用來傳送表單查詢車次的網址
    qResp = requests.post('https://tip.railway.gov.tw' + queryUrl, data=formData)
    qSoup = BeautifulSoup(qResp.text, 'html5lib')
    # tr 代表一列車次的所有資料
    trs = qSoup.find_all('tr', 'trip-column')
    for tr in trs:
        # td 將資料分為一欄一欄的
        td = tr.find_all('td')
        print('%s : %s, %s' % (td[0].ul.li.a.text, td[1].text, td[2].text))
        
getTrip()
```

### 使用 Selenium 抓取回傳後的 HTML，由 Selenium 提供的工具，取得 HTML 裡面的元素值 (使用 XPath)

```python
from bs4 import BeautifulSoup
from selenium import webdriver

try:
    # 讓瀏覽器在背景作業，不要跳出視窗，執行速度比較快
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    # 加入參數 options
    chrome = webdriver.Chrome(options=options, executable_path='/Users/cody8/Documents/chromedriver')
    chrome.set_page_load_timeout(10)
    chrome.get('https://code-gym.github.io/spider_demo/')
    soup = BeautifulSoup(chrome.page_source, 'html5lib')
    print(soup.find('h1').text)

    # 用函式 find_element_by_xpath() 找標題連結元素，並用 click() 點擊
    # XPath 找尋路徑的方式，類似檔案管理中的路徑，從左邊父節點向右邊子節點連接，最左邊的 / 代表根節點的意思
    # div[2] 代表它上面的 body 下有多個標籤 div，而 XPath 的第一個索引值位置是 1，要注意，這裡的 2，就是第二個標籤的意思
    chrome.find_element_by_xpath('/html/body/div[2]/div/div[1]/div[1]/div/div/h3/a').click()
    # 列印找到的內容
    # // 的意思是指，在任何節點下，* 代表所有標籤名稱，屬性 id 為 post-header 的元素
    print(chrome.find_element_by_xpath('//*[@id="post-header"]/div[2]/div/div/h1').text)
    
finally:
    chrome.quit()
```

## ESPN球員球員效率值

[http://insider.espn.com/nba/hollinger/statistics/_/page/1](http://insider.espn.com/nba/hollinger/statistics/_/page/1)

```python
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import random

# 不含頁數
url = 'http://insider.espn.com/nba/hollinger/statistics/_/page/'

try:
    # 用 webdriver 取得一個 Chrome 物件，設定在背景作業，設定 chromedriver 的檔案路徑，設定 timeout 時間
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')
    chrome = webdriver.Chrome(options=options, executable_path='/Users/cody8/Documents/chromedriver')
    chrome.set_page_load_timeout(10)
    
    # 抓取每一頁的資料
    for i in range(1, 9):
        # 加上頁數
        _url = url + str(i)
        # 印出網址，方便觀察
        print(_url)
        # 用 get 取得網頁資訊
        chrome.get(_url)
        # 將網頁內容交給 BeautifulSoup 尋找網頁中的元素
        soup = BeautifulSoup(chrome.page_source, 'html5lib')
        trs = soup.find('tbody').find_all('tr')
        # tr 是每一列資料，td 是每一列中的欄位
        for tr in trs:
            # 將 tr 中的所有 td 資料存在 tds 陣列中
            tds = [td for td in tr.children]
            # 為了避免抓到欄位名稱跟第一列標題
            # 先找出每一列中第一欄的資料，並清除空白字元
            rk = tds[0].text.strip()
            if rk == 'RK' or len(tds) < 2:
                continue
            name = tds[1].text
            per = tds[11].text
            print('%s :%s' % (name, per))
        # 假裝自己不是爬蟲程式，用亂數去讓程式延遲執行
        wait = random.randint(2, 6)
        # 印出等待時間，方便觀察
        print('wait time : %d' % wait)
        time.sleep(wait)
    
finally:
    # 關閉 driver
    chrome.quit()
```

```python
import requests
import json

# API
url = 'https://stats.nba.com/stats/leaguedashplayerstats?College=&Conference=&Country=&DateFrom=&DateTo=&Division=&DraftPick=&DraftYear=&GameScope=&GameSegment=&Height=&LastNGames=0&LeagueID=00&Location=&MeasureType=Base&Month=0&OpponentTeamID=0&Outcome=&PORound=0&PaceAdjust=N&PerMode=PerGame&Period=0&PlayerExperience=&PlayerPosition=&PlusMinus=N&Rank=N&Season=2021-22&SeasonSegment=&SeasonType=Regular+Season&ShotClockRange=&StarterBench=&TeamID=0&TwoWay=0&VsConference=&VsDivision=&Weight='
# 建立自訂的 request header，讓我們更像是瀏覽器，這些 header 資訊告訴主機我們是用哪一種瀏覽器，使用什麼語言等...
# 避免在大量擷取資料的過程被鎖 IP
custom_headers = {
    'Host': 'stats.nba.com',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:72.0) Gecko/20100101 Firefox/72.0',
    'Accept': 'application/json, text/plain, */*',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br',
    'x-nba-stats-origin': 'stats',
    'x-nba-stats-token': 'true',
    'Connection': 'keep-alive',
    'Referer': 'https://stats.nba.com/',
    'Pragma': 'no-cache',
    'Cache-Control': 'no-cache',
}

# 用 requests 連上網站取得回傳資料
resp = requests.get(url, headers=custom_headers)
# 因為是 json 格式的字串，所以用函式 loads 讀取資料
datas = json.loads(resp.text)
players = datas['resultSets'][0]['rowSet']
for player in players:
    print(player[1])
```
